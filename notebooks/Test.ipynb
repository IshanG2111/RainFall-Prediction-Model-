```python
import pandas as pd

grid_df = pd.read_parquet(
    "/Users/saptarshiroy/Desktop/PROJECTS/RainFall-Prediction-Model-/data/grid_definition.parquet"
)

final_df = pd.read_parquet(
    "/Users/saptarshiroy/Desktop/PROJECTS/RainFall-Prediction-Model-/data/final_dataset.parquet"
)

print("Grid file shape:", grid_df.shape)
print("Final dataset shape:", final_df.shape)
```

```python
grid_ids_grid = set(grid_df["grid_id"])
grid_ids_final = set(final_df["grid_id"])

missing = grid_ids_final - grid_ids_grid
extra = grid_ids_grid - grid_ids_final

print("Missing grid_ids in grid_definition:", len(missing))
print("Extra grid_ids in grid_definition:", len(extra))
```

```python
merged = (
    final_df[["grid_id", "lat_center", "lon_center"]]
    .drop_duplicates()
    .merge(
        grid_df[["grid_id", "lat_center", "lon_center"]],
        on="grid_id",
        suffixes=("_final", "_grid")
    )
)

lat_diff = (merged["lat_center_final"] - merged["lat_center_grid"]).abs().max()
lon_diff = (merged["lon_center_final"] - merged["lon_center_grid"]).abs().max()

print("Max latitude difference:", lat_diff)
print("Max longitude difference:", lon_diff)
```

```python
if len(missing) == 0 and len(extra) == 0 and lat_diff == 0 and lon_diff == 0:
    print("✅ grid_definition.parquet is UNCHANGED and SAFE to use")
else:
    print("❌ grid_definition.parquet has CHANGED — STOP and revalidate")
```
