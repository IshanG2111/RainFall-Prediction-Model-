# Rainfall Feature Analysis & Inconsistency Diagnosis

This notebook explores the relationship between meteorological features and
rainfall (`rain_mm`) to understand why city-wise predictions were identical
and how spatial features (latitude & longitude) affect the model.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
```

```python
# Absolute path (safe for notebook)
path = "/Users/saptarshiroy/Desktop/PROJECTS/RAINFALL DETECTION/data/final_dataset.parquet"

df = pd.read_parquet(path)

df.head()
```

```python
print("Shape:", df.shape)
df.info()

df['date'].nunique(), df['date'].unique()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['rain_mm'], bins=50)
plt.xlabel("Rainfall (mm)")
plt.ylabel("Frequency")
plt.title("Distribution of Rainfall (rain_mm)")
plt.show()
```

```python
# X AXIS COORDINATES

FEATURES = [
    'lat_center',
    'lon_center',
    'wind_speed',
    'lst_k',
    'cer',
    'cot',
    'uth',
    'olr',
    'hem',
    'monsoon_flag'
]
```

```python
for col in FEATURES:
    plt.figure(figsize=(6, 4))
    plt.scatter(df[col], df['rain_mm'], alpha=0.3)
    plt.xlabel(col)
    plt.ylabel("rain_mm")
    plt.title(f"rain_mm vs {col}")
    plt.show()
```

```python
plt.figure(figsize=(10, 8))
corr = df[FEATURES + ['rain_mm']].corr()

sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    square=True
)

plt.title("Correlation Matrix with rain_mm")
plt.show()
```

#CITIES GETTING SAME PREDICTION

Reason

```python
# Pick two distant cities
kolkata = df.iloc[(df['lat_center'] - 22.57).abs().idxmin()]
delhi = df.iloc[(df['lat_center'] - 28.61).abs().idxmin()]

kolkata[FEATURES], delhi[FEATURES]
```

```python
daily_df = df.groupby('date').mean(numeric_only=True).reset_index()
daily_df
```

### Key Findings

1. The dataset contains strong spatial density but weak temporal variation.
2. Meteorological features are nearly identical across grid points for a given date.
3. Without latitude and longitude, the regression model collapses to a mean prediction.
4. Including spatial coordinates is necessary to introduce city-level differentiation.
5. Identical predictions are expected behavior, not a model bug.
