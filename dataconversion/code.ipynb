```python
import h5py
import pandas as pd
import numpy as np
import os

def extract_all(h5_file, output_folder="output"):
    # Create output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    with h5py.File(h5_file, 'r') as f:
        def visit(name, obj):
            if isinstance(obj, h5py.Dataset):
                print("Extracting:", name)
                data = np.array(obj)

                # Clean dataset name for file output
                safe_name = name.replace("/", "_")
                csv_name = os.path.join(output_folder, f"{safe_name}.csv")

                # Handle 1D/2D/3D datasets safely
                if data.ndim == 1:
                    df = pd.DataFrame(data)
                else:
                    df = pd.DataFrame(data.reshape(data.shape[0], -1))

                df.to_csv(csv_name, index=False)
                print(f"Saved → {csv_name}")

        f.visititems(visit)

# Example
extract_all("3RIMG_01NOV2025_0015_L2B_HEM_V01R00.h5")
```
